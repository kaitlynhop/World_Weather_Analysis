```python
## Import the dependencies.
# random number generator
import numpy as np
import random
# Retrieve nearest city
from citipy import citipy
# For API Call
import requests
import time
# API key
import sys
sys.path.append('..')
from config import weather_api_key
# Dataframe
import pandas as pd
from datetime import datetime
```

```python
## Generate list of 2000 lat and long values with np and random
lats = np.random.uniform(low=-90, high=90, size=2000)
lngs = np.random.uniform(low=-180, high=180, size=2000)
# zip lat and long values together and store as list
lat_lngs = zip(lats, lngs)
coordinates = list(lat_lngs)
print(coordinates[2])
```

```python
### Generate list of nearest cities with only unique values
# Initialize list to hold cities
cities = []
# loop through coordinates list to get nearest city name
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
 # Add uniques cities to list   
    if city not in cities:
        cities.append(city)

print(len(cities))        

```

```python
### Make API call and gather data
# Initialize list to hold city data
city_data = []

# set counters to keep track
record_count = 1
set_count = 1

# base url for open weather map
base_url = "https://api.openweathermap.org/data/2.5/weather"

```

```python
for i, city in enumerate(cities):
    print(city[6])
```

```python
#Loop through cities list to get city for url code
params = {
        "appid": weather_api_key,
        "units": "imperial"
    }
```

```python
#Loop through cities list to get city for url code
for i, city in enumerate(cities):
    
    #Set trackers
    if (i % 50 == 0 & i >= 50):
        record_count = 1
        set_count += 1
        time.sleep(60)
    
    print(f"Processing record: {record_count} of set: {set_count} | {city}")
    record_count +=1
    # Build API Url for each city
    url_name = city.replace(" ","+")
    params["q"] = url_name
     #Attempt to parse JSON data and retreive desired information
    try:
        weather = requests.get(base_url, params=params).json()
        city_lat = weather['coord']['lat']
        city_lng = weather['coord']['lon']
        city_max_temp = weather['main']['temp_max']
        city_humidity = weather['main']['humidity']
        city_clouds = weather['clouds']['all']
        city_wind = weather['wind']['speed']
        city_description = weather['weather'][0]['description']
        city_country = weather['sys']['country']
        # Attempt to make this freaking data list
        city_data.append({"City Name": city.title(),
                         "Country": city_country,
                         "Lat": city_lat,
                         "Lng": city_lng,
                         "Max Temp": city_max_temp,
                         "Humidity": city_humidity,
                         "Cloudiness": city_clouds,
                         "Wind Speed": city_wind,
                         "Current Descriptions": city_description})
    
    except:
        print(f"City: {city} error")
        print("Current number of cities in list:" + str(len(city_data)))
        pass
          
print("----------------- \n Method Complete \n ----------------")
    
```

```python
## store city data as DF
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
city_data_df.rename({'City Name':'City'}, axis=1, inplace=True)
city_data_df.head(10)
```

```python
## Export DF as CSV
city_data_df.to_csv('WeatherPy_Database.csv', index_label="City_ID")
```
